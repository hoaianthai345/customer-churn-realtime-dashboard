Muc tieu:
- giu nguyen logic nghiep vu cua `feature_prep.ipynb` cho phan label va feature layer;

```python
from pathlib import Path
import sys
import pandas as pd

CANDIDATE_NOTEBOOK_DIRS = [
    Path.cwd(),
    Path.cwd() / 'project-realtime-bi' / 'notebooks',
]

for candidate in CANDIDATE_NOTEBOOK_DIRS:
    helper_path = candidate / 'feature_prep_from_cache_lib.py'
    if helper_path.exists():
        sys.path.insert(0, str(candidate))
        NOTEBOOK_DIR = candidate
        break
else:
    raise FileNotFoundError('Khong tim thay feature_prep_from_cache_lib.py')

from feature_prep_from_cache_lib import run_feature_prep_from_cache

print(f'NOTEBOOK_DIR = {NOTEBOOK_DIR}')
```

```python
OUTPUT_SUBDIR = 'feature_store_from_cache'
STRICT_TEST_COHORT_MATCH = False

result = run_feature_prep_from_cache(
    output_subdir=OUTPUT_SUBDIR,
    strict_test_cohort_match=STRICT_TEST_COHORT_MATCH,
)

summary_df = pd.DataFrame([result['summary']])
coverage_df = result['cache_coverage_df']
submission_report = result['submission_report']

display(summary_df)
display(coverage_df)
if submission_report is not None:
    display(pd.DataFrame([submission_report]))
```
